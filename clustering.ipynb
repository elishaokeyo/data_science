```python
# clustering 
#group data to different called clusters
import pandas 
# import data 
data=pandas.read_csv("data/Mall_Customers.csv")
# read data 
data
```

```python
# check if empty records
data.isnull().sum()
```

```python
# dropping of records with empties
data.dropna(inplace=True)
```

```python
# confirm if empties are dropped 
data.isnull().sum()
```

```python
# split data 
array=data.values
X=array[:, 2:5]
X
```

```python
# fit data to model  
from sklearn.cluster import KMeans
model=KMeans(n_clusters=10, random_state=42)
model.fit(X)
```

```python
# find clusters 
means=model.cluster_centers_
clusters=pandas.DataFrame(means, columns=["Age", "Income", "Score"])
clusters
```

```python
# get specific cluster data 
data["no"]=model.labels_
cluster5=data[data["no"]==5]
cluster5
```

```python
# save to csv 
cluster5.to_csv("cluster5.csv")
```
