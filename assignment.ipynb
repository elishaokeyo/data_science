```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.available
plt.style.use('seaborn-v0_8')
```

```python
data.groupby("category").size()
```

```python
sns.countplot(x=data["category"], palette="magma")
```

```python
sns.scatterplot
colors=["#3949AB", "#2E7D32", "#546E7A"]
sns.scatterplot(x=data["category"], y="unit price", data=data ,palette="Greens", hue="category")
```

```python
sns.barplot
colors=["#3949AB", "#2E7D32", "#546E7A"]
sns.barplot(x=data["category"], y="unit price",data=data, palette="magma", hue="category")
plt.title("unit price in relation to category")
```

```python
sns.swarmplot
colors=["#3949AB", "#2E7D32", "#546E7A"]
sns.swarmplot(x=data["category"], y="ext price",data=data, palette="magma", hue="category")
plt.title("distribution of ext price in relation to category")
```

```python
data.groupby("category").size().plot(kind="pie", autopct="%1.1f%%")
plt.title("distribution of category")                  
```
